```python
import pandas as pd
```

```python
#loading data
blinkit   = pd.read_csv("blinkit_mock_export.csv")
zepto     = pd.read_csv("zepto_mock_export.csv")
instamart = pd.read_csv("instamart_mock_export.csv")
```

```python
#renaming columns for better readabililty and maintain standardized formats

blinkit = blinkit.rename(columns={
    "product_name" : "product_name",
    "units_sold"   : "units_sold",
    "revenue_inr"  : "revenue_inr",
    "date"         : "date",
    "city"         : "city"
})

zepto = zepto.rename(columns={
    "item_name" : "product_name",
    "qty_sold"  : "units_sold",
    "gmv"       : "revenue_inr",
    "date"      : "date",
    "location"  : "city"
})

instamart = instamart.rename(columns={
    "sku_description" : "product_name",
    "quantity"        : "units_sold",
    "sales_value"     : "revenue_inr",
    "date"            : "date",
    "region"          : "city"
})


blinkit.drop(columns=["internal_id"], errors="ignore", inplace = True)
instamart.drop(columns=["remarks"], errors="ignore", inplace = True)
```

```python
# adding platform columns will help in the master CSV
blinkit["platform"]   = "Blinkit"
zepto["platform"]     = "Zepto"
instamart["platform"] = "Instamart"

# making all date formats same, writing current format for my convenience
# Blinkit:   DD-04-2025  
# Zepto:     2025/04/10  
# Instamart: April 05 2025 

blinkit["date"]   = pd.to_datetime(blinkit["date"],   dayfirst=True)
zepto["date"]     = pd.to_datetime(zepto["date"],     yearfirst=True)
instamart["date"] = pd.to_datetime(instamart["date"], format="%B %d %Y")


for df in [blinkit, zepto, instamart]:
    df["date"] = df["date"].dt.strftime("%d-%m-%Y")

# city name changes
city_map = {
    "delhi"     : "Delhi",
    "new delhi" : "Delhi",
    "New Delhi" : "Delhi",
    "bangalore" : "Bengaluru",
    "Bangalore" : "Bengaluru",
    "mumbai"    : "Mumbai",
    "hyderabad" : "Hyderabad",
    "chennai"   : "Chennai"
}

for df in [blinkit, zepto, instamart]:
    df["city"] = df["city"].str.strip().replace(city_map)
```

```python

# product name standardization
product_map = {
    # Lays 
    "Lays Classic Salted 26g"    : "Lays Classic Salted 26g",
    "LAYS CLASSIC SALTED"        : "Lays Classic Salted 26g",
    "lay's classic - 26gm"       : "Lays Classic Salted 26g",
    "lays classic salted"        : "Lays Classic Salted 26g",

    # Kurkure 
    "kurkure masala munch - 90gm": "Kurkure Masala Munch 90g",
    "KURKURE MASALA MUNCH"       : "Kurkure Masala Munch 90g",
    "Kurkure Masala Munch 90g"   : "Kurkure Masala Munch 90g",

    # Oreo 
    "Oreo Original Cream 120g"   : "Oreo Original Cream 120g",
    "OREO ORIGINAL"              : "Oreo Original Cream 120g",
    "oreo cream biscuits 120gm"  : "Oreo Original Cream 120g",

    # Too Yumm 
    "Too Yumm Multigrain Crisps 60g" : "Too Yumm Multigrain Crisps 60g",
    "TOO YUMM MULTIGRAIN"            : "Too Yumm Multigrain Crisps 60g",
    "too yumm crisps 60gm"           : "Too Yumm Multigrain Crisps 60g",

    # Haldirams 
    "Haldirams Aloo Bhujia 200g"     : "Haldirams Aloo Bhujia 200g",
    "HALDIRAMS BHUJIA"               : "Haldirams Aloo Bhujia 200g",
    "haldiram aloo bhujia - 200gm"   : "Haldirams Aloo Bhujia 200g",
}

for df in [blinkit, zepto, instamart]:
    df["product_name"] = df["product_name"].str.strip().replace(product_map)

# DATA QUALIITY HANDLING, LIKE NULL VALUES AND ADDING FLAGS FOR FASTER IDENTIFICATION OF NULL VALUE ROWS
zepto["data_quality_flag"] = zepto["revenue_inr"].isna().map(
    {True: "missing_revenue", False: "ok"}
)
blinkit["data_quality_flag"]   = "ok"
instamart["data_quality_flag"] = instamart["units_sold"].apply(
    lambda x: "zero_quantity" if x == 0 else "ok"
)
```

```python
#now making a standardized master csv
final_cols = ["platform", "product_name", "units_sold",
              "revenue_inr", "date", "city", "data_quality_flag"]

blinkit   = blinkit[final_cols]
zepto     = zepto[final_cols]
instamart = instamart[final_cols]

# ── 9. CONCATENATE AND EXPORT ─────────────────────────────────────────────────
master_df = pd.concat([blinkit, zepto, instamart], ignore_index=True)
master_df["date"] = pd.to_datetime(master_df["date"], format="%d-%m-%Y").dt.strftime("%Y-%m-%d")

master_df.to_csv("unified_output.csv", index=False)
print("\n✓ unified_output.csv saved successfully.")
```
