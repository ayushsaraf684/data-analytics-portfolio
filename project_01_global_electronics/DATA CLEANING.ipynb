## IMPORTING DATA THROUGH PANDAS

```python
import pandas as pd

customers = pd.read_csv('Customers.csv', encoding='latin-1')
            
```

```python
data_dictionary = pd.read_csv('Data_Dictionary.csv', encoding='latin-1')
exchange_rate = pd.read_csv('Exchange_Rates.csv', encoding='latin-1')
products = pd.read_csv('Products.csv', encoding='latin-1')
sales = pd.read_csv('Sales.csv', encoding='latin-1')
stores = pd.read_csv('Stores.csv', encoding='latin-1')
```

## CLEANING DATA TABLE-WISE



### PRODUCTS TABLE

```python
products
products['Unit Cost USD'] = products['Unit Cost USD'].str.replace('$', '')
products['Unit Price USD'] = products['Unit Price USD'].str.replace('$', '')
```

```python
products.rename(columns = {'ProductKey':'product_key',
                           'Product Name' : 'product_name',
                           'Brand' : 'brand',
                           'Color' : 'color',
                           'Unit Cost USD' : 'unit_cost_usd',
                           'Unit Price USD' : 'unit_price_usd',
                           'SubcategoryKey' : 'subcategory_key',
                           'Subcategory' : 'subcategory',
                           'CategoryKey' : 'category_key',
                           'Category' : 'category'
                           }, inplace=True)
```

```python
products['unit_price_usd'] = pd.to_numeric(products['unit_price_usd'], errors='coerce')
products['unit_cost_usd'] = pd.to_numeric(products['unit_cost_usd'], errors='coerce')

# Check the data types
print(products.dtypes)
```

### EXCHANGE RATE TABLE

```python
exchange_rate['Date'] = pd.to_datetime(exchange_rate['Date'])
```

```python
exchange_rate.rename(columns = {
    'Date': 'date',
    'Currency': 'currency',
    'Exchange': 'exchange'
}, inplace=True)
```

### SALES TABLE

```python
sales['Order Date'] = pd.to_datetime(
    sales['Order Date'],
    dayfirst=True,
    errors='coerce'
)
sales['Delivery Date'] = pd.to_datetime(
    sales['Delivery Date'],
    dayfirst=True,
    errors='coerce'
)
sales.rename(columns = {
    'Order Number': 'order_number',
    'Line Item': 'line_item',
    'Order Date': 'order_date',
    'Delivery Date': 'delivery_date',
    'CustomerKey': 'customer_key',
    'StoreKey': 'store_key',
    'ProductKey': 'product_key',
    'Quantity': 'quantity',
    'Currency Code': 'currency_code'
}, inplace=True)
```

```python
sales.info()
```

```python
sales['delivery_date'].isna().sum()
```

### STORES TABLE

```python
stores.rename(columns = {
    'StoreKey': 'store_key',
    'Country': 'country',
    'State': 'state',
    'Square Meters': 'square_meters',
    'Open Date': 'open_date'
}, inplace=True)
```

```python
stores['open_date'] = pd.to_datetime(stores['open_date'], dayfirst = True, errors = 'coerce')
```

```python
stores.head()
```

## CUSTOMERS TABLE

```python
customers.head()
```

```python
customers.rename(columns = {
    'CustomerKey': 'customer_key',
    'Gender': 'gender',
    'Name': 'name',
    'City': 'city',
    'State Code': 'state_code',
    'State': 'state',
    'Zip Code': 'zip_code',
    'Country': 'country',
    'Continent': 'continent',
    'Birthday': 'birthday'
}, inplace=True)
```

```python
customers['birthday'] = pd.to_datetime(customers['birthday'],
                                      dayfirst = True,
                                      errors = 'coerce')
```

# EXPORTING THE DATA TO MS SQL DATABASE FOR E.D.A

```python
# creating the connection, engine creates the blueprint of the connection
# my database name is global_elec
import pandas as pd
from sqlalchemy import create_engine
engine = create_engine("mssql+pyodbc://@localhost\\SQLEXPRESS/global_elec?driver=ODBC+Driver+18+for+SQL+Server&trusted_connection=yes&TrustServerCertificate=yes")
```

```python
customers.to_sql(name='customers', con = engine, index = False)
sales.to_sql(name='sales', con = engine, index = False)
products.to_sql(name='products', con = engine, index = False)
stores.to_sql(name='stores', con = engine, index = False)
data_dictionary.to_sql(name='data_dictionary', con = engine, index = False)
exchange_rate.to_sql(name='exchange_rate', con = engine, index = False)
```
