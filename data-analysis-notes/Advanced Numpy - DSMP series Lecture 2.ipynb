# UNDERSTANDING WHY NUMPY ARRAYS ARE FASTER THAN PYTHON LISTS

## SPEED - ARRAYS ARE WAY FASTER

````python
import numpy as np
import time
import matplotlib.pyplot as plt
````

````python
a = [i for i in range(10000000)]
b = [i for i in range(10000000, 20000000)]
c = []
start = time.time()
for i in range(len(a)):
    c.append(a[i] + b[i])
print(time.time() - start)

#this loop took 0.40 seconds to run
````

````python

a = np.arange(10000000)
b = np.arange(10000000, 20000000)

start = time.time()
c = a + b
print(time.time() - start)

# whereas this loop only takes 0.05 seconds to run
````

## MEMORY - ARRAYS ARE WAY MORE EFFICIENT

````python
import sys
````

````python
a = [i for i in range(10000000)]
print(f' Size in Megabytes of this list "a" is {sys.getsizeof(a)/(1024**2)}')
# this is the size of only the list of references, and not the actual size of list of integer objects
````

````python
b = np.arange(10000000, 20000000, dtype = np.int32)
print(f' Size in Megabytes of this array "b" is {sys.getsizeof(b)/(1024**2)}')
````

## CONVINIENCE - ARRAYS ARE EASIER TO CODE

````python
# To add numbers in 2 lists
a = [i for i in range(10000000)]
b = [i for i in range(10000000, 20000000)]
for i in range(len(a)):
    c.append(a[i] + b[i])
print(c)

# To add numbers in 2 arrays
a = np.arange(10000000)
b = np.arange(10000000, 20000000)
c = a + b
print(c)
````

# ADVANCED INDEXING

## FANCY INDEXING

````python
# They are of 2 types - Fancy Indexing and Boolean Indexing
# Used when there is no pattern to index from an array by manipulating indexes
# for example

b = np.arange(24).reshape(6,4)
````

````python
b
# you want to extract R1C1, R1C2, R2C1, and R2C2 - here there is a pattern, the values are contiguous
# or you want R1C1, R6C1, R1C6 and R6C6 - here also there is a pattern
# but if I say I need R1,R2, R5 and R6 - there is no pattern
# here you need advanced indexing
````

### For fancy indexing, just provide what all rows and columns you need in lists

````python
b[[0,1,4,5]] # generates R1,R2, R5 and R6
````

````python
b[:,[0,2,3]] #generates C1, C3 and C4
````

## BOOLEAN INDEXING

````python
a = np.random.randint(1,100,24).reshape(6,4)
````

````python
a > 50
````

````python
a[a>50] # generates all numbers greater than 50 in your array
````

````python
a[a%2==0]
````

````python
a[(a%2==0) & (a>50)] #generates all numbers greater than 50 and even
````

````python
a[~(a%7==0)]
````

## WORKING WITH MATHEMATICAL FORMULAS

### THESE ARE USER DEFINED FUNCTIONS, NOT PRE BUILT IN PYTHON

````python
#sigmoid
def sigmoid(array):
    return 1 / (1 + np.exp(-array)))

a = np.arange(-10)

sigmoid(a)
````

````python
#mean squared error - a loss function used in ML Models specifically Linear Regression
````

````python
actual = np.random.randint(1,100,34)
predicted = np.random.randint(1,100,34)
````

````python
actual
````

````python
predicted
````

````python
def mse(a,p):
    return np.mean((a - p)**2)
mse(actual,predicted)
````

````python
# binary cross entropy
actual = np.round(np.random.random((20))).astype(np.int32)
predicted = np.round(np.random.random((20)),2)
````

````python
print(actual)
print(predicted)
````

````python
def bce(a,p):
    p = np.clip(p, 1e-7, 1 - 1e-7) 
    return -(np.mean(a * np.log(p) + (1-a) * np.log(1-p)))

bce(actual,predicted)
````
