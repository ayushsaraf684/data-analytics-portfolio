### LIST OF ALL PARAMETERS USED IN PLT.PLOT()

````python
# plt.plot() Parameter Cheatsheet

## color / c -> sets the color of the line.
### hex code
#### RGB tuple (0-1 scale)
#### RGBA tuple
#### matplotlib picks default color

## linestyle / ls -> sets the pattern of the line
#### '-'    or 'solid'
#### '--'   or 'dashed'
#### '-.'   or 'dashdot'
#### ':'    or 'dotted'
#### 'None', ' ', '' ->  no line

## linewidth / lw -> sets the thickness of the line
#### 0.5, 1, 2, 3, 5 -> any positive float/int

## marker -> adds a symbol at each data point.
# '.'   # point
# ','   # pixel
# 'o'   # circle
# 'v'   # triangle_down
# '^'   # triangle_up
# '<'   # triangle_left
# '>'   # triangle_right
# '1'   # tri_down
# '2'   # tri_up
# '3'   # tri_left
# '4'   # tri_right
# 's'   # square
# 'p'   # pentagon
# '*'   # star
# 'h'   # hexagon1
# 'H'   # hexagon2
# '+'   # plus
# 'x'   # x
# 'D'   # diamond
# 'd'   # thin_diamond
# '|'   # vline
# '_'   # hline
# None, '', ' '             # no marker

## markersize / ms -> sets the size of the marker.
# 2, 4, 6, 8, 10, 12  -> any positive float/int

## markerfacecolor / mfc (sets the fill color inside the marker) AND 
## markeredgecolor / mec (sets the border color of the marker)
# Any color name or Hex Code         
# 'none' -  hollow marker
# None - matches line color


## markeredgewidth / mew -> sets the thickness of the marker border.
#### 0.5, 1, 2, 3 - any positive float/int

## alpha -> sets the transparency of the line or marker
#### Ranges 0 to 1, corresponding from TRANSPARENT to OPAQUE

## label -> gives the plot a name for the legend.
#### 'Sales', 'Revenue', 'Profit', None -> any string, used in legend

## drawstyle/ds  -> changes how lines are drawn between points, such as steps
#### controls how points are connected (straight lines vs. step-like jumps, useful for discrete/stepped data like inventory counts).
#### 'default' - straight lines between points (normal)
#### 'steps' - same as 'steps-pre'
#### 'steps-pre' - step happens before the point
#### 'steps-mid' - step happens at the midpoint between points
#### 'steps-post' - step happens after the point ####

## zorder -> controls which plot appears on top when plots overlap
# 0, 1, 2, 3, 10             # higher = drawn on top

## markevery -> shows markers only at selected data points.
# None
# 2                          # every 2nd point
# (0, 2)                     # start at 0, every 2nd point
# slice(0, 10, 2)
# [0, 2, 4, 6]               # specific indices
# 0.1                        # fraction-based spacing
# (0.1, 0.2)                 # fraction-based start + spacing


## visible  ->  Shows or hides the plotted line.
# True    # show the line
# False   # hide the line


## Priority to memorize first
# color / c, linestyle / ls, linewidth / lw, marker, markersize / ms,
# markerfacecolor / mfc, markeredgecolor / mec, markeredgewidth / mew,
# alpha, label
````

## WORKING WITH PARAMETERS OF PLT.SCATTER()

| Parameter | Meaning |
|---|---|
| `x` | X-axis values |
| `y` | Y-axis values |
| `s` | Size of each dot |
| `c` | Color of each dot |
| `cmap` | Color palette for numeric colors |
| `alpha` | Transparency of dots |
| `marker` | Shape of dots |
| `edgecolors` | Border color of dots |
| `linewidths` | Border thickness |
| `label` | Name for legend |


### Important use cases

- **Basic relationship between two variables** — to check correlation between two numerical variables.
- **Bubble chart** — to represent a third variable using dot size.
- **Color mapping** — to represent a third variable using color intensity.
- **Bubble chart with both size and color** — to encode four variables in a single plot.
- **Transparency for overlapping points** — to make dense clusters easier to see.
- **Custom marker shapes** — to differentiate categories visually.
- **Highlighting points with borders** — to make points stand out clearly.

### Important colormaps

| Colormap | Use case |
|----------|----------|
| `viridis` | Best default, colorblind-friendly |
| `plasma` | Bright, high contrast |
| `coolwarm` | Negative vs positive values |
| `Blues` | Single-color intensity |
| `Reds` | Single-color intensity |

### What to look for in a scatter plot

- **Positive correlation** — points rise from left to right.
- **Negative correlation** — points fall from left to right.
- **No correlation** — points are randomly scattered.
- **Clusters** — groups of points form separate patterns.
- **Outliers** — points are far away from the main cluster.

### The one pattern to memorize

```python
plt.scatter(
```

### Important use cases
- **Category comparison** — comparing values across discrete categories (e.g. sales by region).
- **Grouped bar chart** — comparing multiple sub-groups per category (requires manual position shifting, since matplotlib has no built-in grouping).
- **Stacked bar chart** — using `bottom` to stack one bar on top of another to show part-to-whole relationships.
- **Horizontal bars (`plt.barh()`)** — useful when category names are long and don't fit well on the x-axis.
- **Error bars** — using `yerr`/`xerr` to show variability or confidence intervals on top of bar values.
- **Highlighting specific bars** — using a list of colors instead of one color to draw attention to particular bars.

### What to look for in a bar chart
- **Tallest/shortest bars** — quick visual comparison of magnitude across categories.
- **Uneven spacing/grouping** — check if grouped bars are positioned correctly relative to their category.
- **Stacked proportions** — in stacked bars, look at how much each segment contributes to the total.
- **Outlier categories** — bars that are drastically higher/lower than the rest.

## Grouped Bar Chart — Steps (matplotlib has no direct function for this)

### Why this manipulation is needed
matplotlib's `plt.bar()` only plots bars at given x-positions.
To show multiple bars per category side-by-side, you must manually
calculate shifted numeric positions for each group.

---

## Case 1: Two bars per category (e.g. boys vs girls)

1. Convert x-labels to numeric positions using `np.arange(len(x))` → this becomes `bar1`.
2. Plot the first group directly at `bar1`.
3. Create `bar2` by shifting `bar1` to the right: `bar2 = [i + w for i in bar1]`
   (subtract `w` instead if you want it on the left).
4. Plot the second group at `bar2`.
5. Set `plt.xticks()` using the **actual category names**, not bar1 (numbers).
6. Shift the tick positions to sit **between** the two bars:
   `plt.xticks(bar1 + w/2, x)` (use `- w/2` if bar2 was placed on the left).

---

## Case 2: Three bars per category (e.g. gold, silver, bronze)

1. Convert x-labels to numeric positions → `bar1` (this stays the **middle** bar).
2. Create `bar2 = bar1 + w` → right-side bar.
3. Create `bar3 = bar1 - w` → left-side bar.
4. Plot all three bars at `bar1`, `bar2`, `bar3` respectively.
5. Set `plt.xticks(bar1, x)` — **no shifting needed**, since bar1 is already centered
   (it has one bar on each side).

---

## Core logic to remember
- Numeric x-axis positions are required to control spacing/position of each bar.
- `+ width` → shift right, `- width` → shift left.
- Ticks must always be relabeled with the real category names (`x`), since the plotted positions are just numbers.
- Tick position correction (`+ w/2` or `- w/2`) is only needed when the "anchor" bar (bar1) is not naturally centered between the groups — i.e., in the 2-bar case. In the 3-bar case, bar1 is already the center, so no correction is needed.

````python
w=0.4
x = ["CSE", "ISE", "ECE", "EEE", "Mech"]
boys = [100,40,60,240,300]
girls = [200,70,30,120,15]

bar1 = np.arange(len(x))
bar2 = [i + w for i in bar1]

plt.bar(bar1,boys,w,label="boys")
plt.bar(bar2,girls,w,label="girls")
plt.xticks(bar1 + w/2, x)
plt.xlabel("Courses")
plt.ylabel("Students")
plt.title("students Vs Courses")
plt.legend()
plt.show()
````

````python
w=0.2
x = [ "USA", "Russia", "GBR", "China", "Germany" ]
gold = [20,10,6,24,30]
silver = [100,50,30,120,150]
bronze= [200,100,60,240,300]
bar1 = np.arange(len(x))
bar2 = [i + w for i in bar1]
bar3 = [i - w for i in bar1]

plt.bar(bar2,gold,w,label="gold", color = "#d4af37")
plt.bar(bar1,silver,w,label="silver",  color = "#c0c0c0")
plt.bar(bar3,bronze,w,label="bronze",  color = "#c79b56")

plt.xticks(bar1, x)
plt.xlabel("Countries")
plt.ylabel("Medals")
plt.title("Medals Vs Countries")
plt.legend()
plt.show()

````

## plt.hist() Parameter Cheatsheet

| Parameter | Meaning | Accepted values |
|---|---|---|
| `x` | Data to bin and plot | array/list/Series of numbers |
| `bins` | Number of bins or bin edges | integer (e.g. `10`, `20`), or list of edges `[0,10,20,30]`, or `'auto'` |
| `range` | Lower and upper range of bins | tuple of two numbers, e.g. `(0, 100)` |
| `density` | Normalize histogram to show probability instead of counts | `True`, `False` |
| `cumulative` | Show cumulative distribution | `True`, `False` |
| `color` | Fill color of bars | colors or hex codes |
| `edgecolor` / `ec` | Border color of bars | colors or hex codes |
| `linewidth` / `lw` | Border thickness of bars | any positive float/int, e.g. `0.5, 1, 2` |
| `alpha` | Transparency of bars | `0` to `1` |
| `histtype` | Style of histogram | `'bar'`, `'barstacked'`, `'step'`, `'stepfilled'` |
| `align` | Alignment of bars relative to bin edges | `'left'`, `'mid'`, `'right'` |
| `orientation` | Direction of histogram | `'vertical'`, `'horizontal'` |
| `rwidth` | Relative width of bars (gap between bars) | `0` to `1` (e.g. `0.8`) |
| `label` | Name for legend | any string |
| `stacked` | Stack multiple datasets on top of each other | `True`, `False` |

### Important use cases
- **Distribution shape** — see if data is normal, skewed, bimodal, etc.
- **Frequency counts** — how many values fall into each range/bin.
- **Comparing groups** — plot multiple histograms with `alpha` for overlap visibility.
- **Cumulative view** — see running total of data up to each bin.
- **Probability density** — using `density=True` to compare distributions of different sizes.

### What to look for in a histogram
- **Symmetric/bell-shaped** — normal distribution.
- **Right-skewed** — long tail on the right (most values small, few large).
- **Left-skewed** — long tail on the left (most values large, few small).
- **Bimodal** — two peaks, suggests two underlying groups.
- **Uniform** — roughly flat, all bins similar height.
- **Outliers** — isolated bars far from the main cluster.
- `density=False` → y-axis shows how many values fall in each bin.
- `density=True` → y-axis is rescaled so the **area** of all bars sums to 1. Useful when comparing two datasets of different sizes (e.g. 100 vs 10,000 samples) on the same scale.

|---|---|---|
| `x` | Wedge sizes (values to plot) | List/array of positive numbers |
| `labels` | Text label for each wedge | List of strings |
| `colors` | Color of each wedge | Colors or hex codes (list, one per wedge) |
| `autopct` | Format for displaying percentage on wedges | `'%1.1f%%'`, `'%1.0f%%'`, `'%.2f%%'`, or a function |
| `explode` | How far each wedge is pulled out from center | List of floats, `0` to `1` (0 = no offset, 0.1 = slightly out) |
| `shadow` | Adds a shadow beneath the pie | `True`, `False` |
| `startangle` | Angle at which the first wedge starts (counterclockwise from x-axis) | `0` to `360` (commonly `90`) |
| `radius` | Size of the pie | Any positive float (default `1`) |
| `labeldistance` | Distance of labels from center | Float, typically `0` to `1.5` (`1.1` = just outside pie) |
| `pctdistance` | Distance of percentage text from center | Float, typically `0` to `1` (default `0.6`) |
| `counterclock` | Direction wedges are drawn | `True` (counterclockwise), `False` (clockwise) |
| `wedgeprops` | Style properties of wedges (border, width) | Dict, e.g. `{'edgecolor': 'black', 'linewidth': 1}` |
| `textprops` | Style properties of label/percent text | Dict, e.g. `{'fontsize': 10, 'color': 'black'}` |
| `center` | Center position of the pie | Tuple `(x, y)`, default `(0, 0)` |
| `frame` | Whether to draw the axes frame around pie | `True`, `False` |
| `rotatelabels` | Rotate labels to align with wedge angle | `True`, `False` |

### Important use cases
- **Basic proportion breakdown** — show how parts contribute to a whole.
- **Exploded pie** — highlight one or more specific categories.
- **Donut chart** — set `wedgeprops={'width': 0.4}` to create a ring instead of a full pie.
- **Percentage labeling** — use `autopct` to show exact percentage contribution per wedge.
- **Custom color-coded categories** — use `colors` to visually group or brand-match wedges.

### What to look for in a pie chart
- **Dominant category** — largest wedge indicates majority share.
- **Balanced distribution** — wedges of similar size mean no single category dominates.
- **Small/negligible categories** — tiny wedges may be better grouped as "Other".
- **Too many categories** — pie charts become unreadable beyond ~6-7 wedges; consider a bar chart instead.

### The one pattern to memorize
```python
plt.pie(
    sizes,                              # Wedge values
    labels=labels,                      # Wedge names
    colors=['#66b3ff','#99ff99','#ff9999'],  # Wedge colors
    autopct='%1.1f%%',                  # Show percentages
    startangle=90,                      # Rotate starting point
    explode=[0.1, 0, 0],                # Pull out first wedge
    shadow=True,                        # Add shadow
    wedgeprops={'edgecolor': 'black'}   # Wedge border
)
plt.title('Category Distribution')
plt.show()
```
