# Most useful parameters for `plt.xlabel()`, `plt.ylabel()`, and `plt.title()`


### Most useful ones to remember
- `fontsize` → changes text size
- `color` → changes text color
- `fontweight`='bold' → makes text bold
- `labelpad` → adds space between axis label and axis
- `pad` → adds space between title and plot
- `loc` → positions title: 'left', 'center', or 'right'

**For axis labels, labelpad is used. For titles, the equivalent spacing parameter is pad.**


### Text and font parameters

| Parameter                     | Accepted values                                                                               |
| ----------------------------- | --------------------------------------------------------------------------------------------- |
| `xlabel` / `ylabel` / `label` | Any string, e.g. `'Sepal Length'`                                                             |
| `fontsize`                    | Integer or float, e.g. `8`, `12`, `20`                                                        |
| `color`                       | Color names like `'red'`, `'blue'`, `'green'`, hex codes like `'#FF5733'`, or RGB/RGBA tuples |
| `fontweight`                  | `'normal'`, `'bold'`, `'heavy'`, `'light'`, `'ultralight'`                                    |
| `fontstyle`                   | `'normal'`, `'italic'`, `'oblique'`                                                           |
| `fontfamily`                  | `'serif'`, `'sans-serif'`, `'monospace'`, `'cursive'`, `'fantasy'`                            |

### Spacing and alignment parameters

| Parameter                   | Accepted values                               |
| --------------------------- | --------------------------------------------- |
| `labelpad`                  | Any integer or float, e.g. `0`, `5`, `15`     |
| `pad` (title only)          | Any integer or float, e.g. `5`, `10`, `20`    |
| `loc` (title only)          | `'left'`, `'center'`, `'right'`               |
| `rotation`                  | Angle in degrees, e.g. `0`, `45`, `90`        |
| `ha` (horizontal alignment) | `'left'`, `'center'`, `'right'`               |
| `va` (vertical alignment)   | `'top'`, `'center'`, `'bottom'`, `'baseline'` |

### Example with accepted values

```python
plt.xlabel(
    'Sepal Length',
    fontsize=14,
    color='blue',
    fontweight='bold',
    labelpad=10
)

plt.ylabel(
    'Petal Length',
    fontsize=14,
    color='#FF5733',
    fontstyle='italic',
    labelpad=12
)

plt.title(
    'Sepal Length vs Petal Length',
    fontsize=18,
    color='purple',
    fontweight='heavy',
    pad=15,
    loc='center'
)
```



# `plt.text()` Parameters



| Parameter | Meaning | Accepted values |
|---|---|---|
| `x` | X-position of the text | Any float/int (data coordinates) |
| `y` | Y-position of the text | Any float/int (data coordinates) |
| `s` | The text string to display | Any string |
| `fontsize` | Size of the text | Any positive float/int (e.g. 8, 10, 12, 14, 20) |
| `color` / `c` | Color of the text | Colors or hex codes (e.g. `'red'`, `'k'`, `'#FF5733'`) |
| `fontweight` | Boldness of the text | `'normal'`, `'bold'`, `'light'`, `'heavy'`, `100`–`1000` |
| `fontstyle` | Slant of the text | `'normal'`, `'italic'`, `'oblique'` |
| `ha` (horizontalalignment) | Horizontal anchor point of text relative to (x, y) | `'left'`, `'center'`, `'right'` |
| `va` (verticalalignment) | Vertical anchor point of text relative to (x, y) | `'top'`, `'center'`, `'bottom'`, `'baseline'` |
| `rotation` | Angle of the text | Any float/int (degrees, e.g. 0, 45, 90) or `'horizontal'`, `'vertical'` |
| `bbox` | Box drawn around the text | Dict, e.g. `dict(facecolor='yellow', edgecolor='black', boxstyle='round')` |
| `alpha` | Transparency of the text | `0` (transparent) to `1` (opaque) |
| `family` | Font family | `'serif'`, `'sans-serif'`, `'monospace'` |
| `zorder` | Stacking order (drawn above/below other elements) | Any float/int; higher = on top |

### Important use cases
- **Annotating specific points** — labeling a data point directly on the chart (e.g. marking the max value).
- **Adding value labels on bars** — writing the exact number on top of each bar.
- **Custom titles/notes inside the plot** — placing extra context anywhere on the figure, not just the default title area.
- **Highlighting thresholds** — writing text near a horizontal/vertical reference line (e.g. "Target: 100").
- **Styling with bbox** — putting a colored box behind text so it stands out from the background.

### The one pattern to memorize
```python
plt.text(
    x=5,                    # X-position
    y=200,                  # Y-position
    s='Peak Value',         # Text to display
    fontsize=12,
    color='black',
    fontweight='bold',
    ha='center',             # Horizontal alignment
    va='bottom',             # Vertical alignment
    bbox=dict(facecolor='yellow', alpha=0.5, edgecolor='black')
)
```

````python
plt.scatter(iris['SepalLengthCm'], iris['PetalLengthCm'])
plt.xlabel('Sepal Length')
plt.ylabel('Petal Length')
# Here since only 3 species are there then we can assign them with colors to have a 3D graph
````

````python
species_map = {
    'Iris-setosa': 0,
    'Iris-versicolor': 1,
    'Iris-virginica': 2
}

iris['species_code'] = iris['Species'].map(species_map)
````

````python
# here is how you can colour code a particular category
plt.scatter(iris['SepalLengthCm'], iris['PetalLengthCm'], c = iris['species_code'], cmap = "winter")
plt.xlabel('Sepal Length')
plt.ylabel('Petal Length')
plt.colorbar()
plt.title("Sepal Length vs Petal Length")
````

````python
plt.figure(figsize = (10,5)) # this changes the size of the graph

plt.scatter(iris['SepalLengthCm'], iris['PetalLengthCm'], c = iris['species_code'], cmap = "winter")

plt.xlabel(
    'Sepal Length',
    fontsize=14,
    color='blue',
    fontweight='bold',
    labelpad=10
)

plt.ylabel(
    'Petal Length',
    fontsize=14,
    color='#FF5733',
    fontstyle='italic',
    labelpad=12
)

plt.title(
    'Sepal Length vs Petal Length',
    fontsize=18,
    color='purple',
    fontweight='heavy',
    pad=10,
    loc='center'
)
plt.colorbar()
plt.show()
````

## SUBPLOTS

````python
plt.subplots()
# <Figure size 640x480 with 1 Axes>
# the subplot returns 2 objects - a figure (area where graph is plotted) and an axis (X and Y)

fig, ax = plt.subplots()
# we store this in 2 variables and do changes in each one of them as required to generate multiple charts at once
````

````python
# some differences between plt.scatter and plt.subplots

fig,ax = plt.subplots(figsize=(15,6))                                       # we used plt.figure(figsize=(a,b))

ax.scatter(batters['avg'],batters['strike_rate'],color='red',marker='+')    # we used plt.scatter, but here we change the axis
ax.set_title('Something')                                                   # we used plt.title()
ax.set_xlabel('Avg')                                                        # we used plt.xlabel()
ax.set_ylabel('Strike Rate')                                                # we use plt.ylabel()
fig.show()                                                                  # use plt.show()        
````

````python
fig, ax = plt.subplots(nrows = 2, ncols = 1, sharex = True, figsize = (15,10)) 
# sharex = true, makes both the graphs have same x axis

ax[0].scatter(batters['avg'],batters['strike_rate'],color='red')
ax[1].scatter(batters['avg'],batters['runs'])

ax[0].set_title('Avg Vs Strike Rate')
ax[0].set_ylabel('Strike Rate')




ax[1].set_title('Avg Vs Runs')
ax[1].set_ylabel('Runs')
ax[1].set_xlabel('Avg')
````

````python
fig, ax = plt.subplots(nrows=2, ncols=2, figsize=(15,10))


ax[0,0].scatter(batters['avg'], batters['strike_rate'], color='red')
ax[0,1].scatter(batters['avg'], batters['runs'])


ax[1,0].hist(batters['avg'], bins=[0,10,20,30,40,50,60])
ax[1,1].hist(batters['strike_rate'], bins=[0,50,100,150,200,250,300,350,400], log = True)


ax[0,0].set_title('Avg vs Strike Rate')
ax[0,0].set_ylabel('Strike Rate')
ax[0,0].set_xlabel('Avg')

ax[0,1].set_title('Avg vs Runs')
ax[0,1].set_ylabel('Runs')
ax[0,1].set_xlabel('Avg')


ax[1,0].set_title('Distribution of Batting Average')
ax[1,0].set_xlabel('Batting Average')
ax[1,0].set_ylabel('Frequency')

ax[1,1].set_title('Distribution of Strike Rate')
ax[1,1].set_xlabel('Strike Rate')
ax[1,1].set_ylabel('Frequency')

plt.tight_layout()
plt.show()
````
